- Compuesto
> $$C_k=C_0(1+i)^k.$$

- Considere diferentes escenarios de interés. Usted elija.  **(10 puntos)**

# 1)

```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline

```

```python
def interess(C_k, C_0, i):
    return round((C_k/C_0 - 1)/i)
```

```python
interess(100000, 10000, .05)
```

```python
def intcomp(C_k, C_0, i):
    return int(np.log(C_k/C_0)/np.log(1 + i)) + 1
```

```python
intcomp(100000, 10000, .05)
```

# 2)

```python
meta = 100000
interes = np.arange(.05, .15, .01)
```

```python
for inte in interes:
    plt.plot(intcomp(100000, 10000, inte), meta, 'o')
plt.ylabel('Meta')    
plt.xlabel('Meses')
plt.show()
```
